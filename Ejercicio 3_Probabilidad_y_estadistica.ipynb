# Lista de calificaciones

```python
from statistics import*

#lista de calificaciones del grupo a de probabilidad y estadistica

grupoa = [74, 80,65,90,100,95,45,80,73,20,100,96]
print(grupoa)

[70,70,65,90,100,95,45,80,73,20,100,96]

#La funcion type nos devuelve el tipo de dato de la variable
type(grupoa)

#sacar la media de las calificaciones con una formula

promedio = sum(grupoa) / len(grupoa)
print(promedio)

#utilizar la funcion mean que esta en la libreria statistics para sacar
#el promedio

promedio2 = mean(grupoa)
print(promedio2)

#Determine el valor de la mediana con la funcion median
mediana = median(grupoa)
print(mediana)

#Dtermine el valor de la moda con la funcion mode

moda = mode(grupoa)
print(moda)
```

# Fabrica 1

```python
# Librería para generar numeros aleatorios
import random


fabrica1 = []
for i in range(100):
    numero_aleatorio = round(random.uniform(50, 100), 2)
    fabrica1.append(numero_aleatorio)

print(fabrica1)
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

print("Valores de la fabrica 1")
print("La media es= ",media1)
print("La medriana es= ",mediana1)
print("La moda es =",moda1)
```


# **Practica 3. Máximo, mínimo, rango y dispersión de datos.**

```python
import random
import matplotlib.pyplot as plt
from statistics import *


fabrica1 = []
for i in range(100):
    numero_aleatorio = round(random.uniform(50, 100), 2)
    fabrica1.append(numero_aleatorio)

print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
  numero_aleatorio = round(random.uniform(50, 100), 2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
maximo1= max(fabrica1)
minimo1 = min(fabrica1)
rango1 = maximo1 - minimo1
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

print("Estadistica descriptiva de la Fabrica 1 \n", "Media =", media1, "\n Mediana =", mediana1, "\n Moda", moda1, "\n Maximo = ", maximo1, "\n Minimo =", minimo1, "\n Rango", rango1)

```

```python
maximo2 = max(fabrica2)
minimo2 =min(fabrica2)
rango2 = maximo2 - minimo2
media2 = mean (fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)

print("Estadistica descrptiva de la fabrica 2 \n","media = ",media2,
      "\n Mediana =",mediana2, "\n Moda =",moda2, "\n Maximo =", maximo2,
      "\n Minimo =",minimo2,"\n Rango =",rango2)
```

Dispersion de los Datos

```python
desviacion_estandar_fabrica1 = stdev(fabrica1)
print(desviacion_estandar_fabrica1)
```

```python
desviacion_estandar_fabrica2 = stdev(fabrica2)
print(desviacion_estandar_fabrica2)
```

```python
plt.figure(figsize = (8,6))
plt.hist(fabrica1, bins=10, alpha=0.2, color='b', edgecolor = "black")
plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label=f'Media={media1:.2f}')
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={mediana1:.2f}')
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label=f'Moda={moda1:.2f}')
plt.axvline(media1 - desviacion_estandar_fabrica1, color='black', linestyle='dashed', linewidth=2, label=f'Desv. Estándar ={desviacion_estandar_fabrica1:.2f}')
plt.axvline(media1 + desviacion_estandar_fabrica1, color='black', linestyle='dashed', linewidth=2)
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Producción diaria de la fabrica 1")
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.hist(fabrica2, bins=10, alpha=0.2, color='b', edgecolor = "black")
plt.axvline(media2, color='b', linestyle='dashed', linewidth=2, label=f'Media={media1:.2f}')
plt.axvline(mediana2, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={mediana1:.2f}')
plt.axvline(moda2, color='y', linestyle='dashed', linewidth=2, label=f'Moda={moda1:.2f}')
plt.axvline(media2 - desviacion_estandar_fabrica2, color='black', linestyle='dashed', linewidth=2, label=f'Desv. Estándar ={desviacion_estandar_fabrica2:.2f}')
plt.axvline(media2 + desviacion_estandar_fabrica2, color='black', linestyle='dashed', linewidth=2)
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Producción diaria de la fabrica 2")
plt.legend()
plt.show()
```
