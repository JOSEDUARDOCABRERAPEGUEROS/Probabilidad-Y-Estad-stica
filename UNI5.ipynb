```python
import pandas as pd
df1 = pd.read_csv("grupo1.csv")
display(df1.head())
df2 = pd.read_csv("grupo2.csv")
display(df2.head())
df3 = pd.read_csv("grupo3.csv")
display(df3.head())
```

```python
import matplotlib.pyplot as plt

print("Estadísticas Descriptivas para Grupo 1:")
print(df1[["HorasEstudio", "Calificacion"]].describe())
print("\n")

plt.figure(figsize=(10, 5))
plt.hist(df1["Calificacion"], bins=10, edgecolor="black")
plt.title("Histograma de Calificaciones para Grupo 1")
plt.xlabel("Calificación Final")
plt.ylabel("Frecuencia")
plt.show()
```

```python
print("Estadísticas Descriptivas para Grupo 2:\n")
print(df2[["HorasEstudio", "Calificacion"]].describe())
print("\n")

plt.figure(figsize=(10, 5))
plt.hist(df2["Calificacion"], bins=10, edgecolor="black")
plt.title("Histograma de calificaciones para Grupo 2")
plt.xlabel("Calificación Final")
plt.ylabel("Frecuencia")
plt.show()
```

```python
print("Estadísticas Descriptivas para Grupo 3:\n")
print(df3[["HorasEstudio", "Calificacion"]].describe())
print("\n")

plt.figure(figsize=(10, 5))
plt.hist(df3["Calificacion"], bins=10, edgecolor="black")
plt.title("Histograma de calificaciones para Grupo 3")
plt.xlabel("Calificación Final")
plt.ylabel("Frecuencia")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
#Grupo 1
X_g1 = df1[['HorasEstudio']]
y_g1 = df1['Calificacion']

# Entrenar el modelo de regresión lineal
modelo_rl_g1 = LinearRegression()
modelo_rl_g1.fit(X_g1, y_g1)

# Obtener la pendiente e intercepto
pendiente_g1 = modelo_rl_g1.coef_[0]
intercepto_g1 = modelo_rl_g1.intercept_

print("\nModelo de Regresión Lineal Grupo 1")
print(f"Ecuación de la línea de regresión: Calificacion = {pendiente_g1:.2f} * HorasEstudio + {intercepto_g1:.2f}")

# Calcular el coeficiente de correlación de Pearson
correlacion_g1, _ = pearsonr(df1['HorasEstudio'], df1['Calificacion'])
print(f"Coeficiente de correlación de Pearson (r): {correlacion_g1:.2f}")
```

```python
#Grupo 2
X_g2 = df2[['HorasEstudio']]
y_g2 = df2['Calificacion']

# Entrenar el modelo de regresión lineal
modelo_rl_g2 = LinearRegression()
modelo_rl_g2.fit(X_g2, y_g2)

# Obtener la pendiente e intercepto
pendiente_g2 = modelo_rl_g2.coef_[0]
intercepto_g2 = modelo_rl_g2.intercept_

print("\nModelo de Regresión Lineal Grupo 2")
print(f"Ecuación de la línea de regresión: Calificacion = {pendiente_g2:.2f} * HorasEstudio + {intercepto_g2:.2f}")

# Calcular el coeficiente de correlación de Pearson
correlacion_g2, _ = pearsonr(df2['HorasEstudio'], df2['Calificacion'])
print(f"Coeficiente de correlación de Pearson (r): {correlacion_g2:.2f}")
```

```python
#Grupo 3
X_g3 = df3[['HorasEstudio']]
y_g3 = df3['Calificacion']

# Entrenar el modelo de regresión lineal
modelo_rl_g3 = LinearRegression()
modelo_rl_g3.fit(X_g3, y_g3)

# Obtener la pendiente e intercepto
pendiente_g3 = modelo_rl_g3.coef_[0]
intercepto_g3 = modelo_rl_g3.intercept_

print("\nModelo de Regresión Lineal Grupo 3")
print(f"Ecuación de la línea de regresión: Calificacion = {pendiente_g3:.2f} * HorasEstudio + {intercepto_g3:.2f}")

# Calcular el coeficiente de correlación de Pearson
correlacion_g3, _ = pearsonr(df3['HorasEstudio'], df3['Calificacion'])
print(f"Coeficiente de correlación de Pearson (r): {correlacion_g3:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.scatterplot(x='HorasEstudio', y='Calificacion', data=df1, label='Datos del Grupo 1')
plt.plot(df1['HorasEstudio'], modelo_rl_g1.predict(df1[['HorasEstudio']]), color='red', label=f'Regresión Lineal (y={pendiente_g1:.2f}x + {intercepto_g1:.2f})')
plt.title('Relación entre Horas de Estudio y Calificación - Grupo 1')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación Final')
plt.legend()
plt.grid(True)
plt.show()
```
