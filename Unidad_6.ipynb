2.-Realizar un Histograma de precios de los automviles

```python
import math
```

```python
df = pd.DataFrame({'Precios': autos.price})
df
```

```python
# Calcular el nmero de intervalos utilizando la regla de Sturges
n = len(autos.price)
k = int( 1 + math.log2(100))
print("Valor de n = ", n , " \nNmero de intervalos =", k)
```

```python
# Crear la tabla de frecuencias

tabla_frecuencia = pd.cut( df['Precios'], bins=k, include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns = ['Clase', 'Frec']
tabla_frecuencia
```

```python
# Ordenar la tabla de frecuencia por el intervalo
tabla_frecuencia = tabla_frecuencia.sort_values(by='Clase')
tabla_frecuencia
```

```python
# Generar columnas siguientes
tabla_frecuencia['Frec Rel'] = tabla_frecuencia['Frec'] / n
tabla_frecuencia['Frec Porc'] = tabla_frecuencia['Frec'] / n * 100
tabla_frecuencia['Frec Ac'] = tabla_frecuencia['Frec'].cumsum()
tabla_frecuencia['Frec Rel Ac'] = tabla_frecuencia['Frec Rel'].cumsum()
tabla_frecuencia['Frec Porc Ac'] = tabla_frecuencia['Frec Porc'].cumsum()

# Mostrar la tabla de frecuencia
print(tabla_frecuencia)
```

```python
# Crear un histograma basado en la tabla de frecuencia
plt.hist(df['Precios'], bins=k, alpha = 0.5)
plt.title('Histograma del precio de los autos basado en la regla de Sturges')
plt.xlabel('Precios')
plt.ylabel('Autos')
```

Conclusiones: La distribucin de los precios de los autos en el histograma sigue una tendencia sesgada hacia la derecha, lo que indica que la mayora de los autos tienen precios ms bajos (entre
15,000). A medida que el precio aumenta, la frecuencia de autos en esos rangos disminuye drsticamente, lo que sugiere que los autos ms caros son menos comunes en el conjunto de datos. Esta distribucin es tpica en mercados donde los autos ms accesibles tienen una mayor demanda y disponibilidad en comparacin con los modelos ms costosos.

Adems, el tipo sedan muestra una gran cantidad de valores atpicos (outliers), lo que indica que existen algunos modelos significativamente ms caros que la mayora. Esto sugiere que los autos deportivos y de lujo (como convertibles y hardtops) tienden a tener un rango de precios ms amplio y pueden alcanzar valores ms elevados.
