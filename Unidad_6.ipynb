```python
# Importar libreras necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el dataset (Asegrate de tener el archivo en tu entorno de Google Colab o usar una URL)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
columns = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors", "body-style", "drive-wheels",
           "engine-location", "wheel-base", "length", "width", "height", "curb-weight", "engine-type", "num-of-cylinders",
           "engine-size", "fuel-system", "bore", "stroke", "compression-ratio", "horsepower", "peak-rpm", "city-mpg",
           "highway-mpg", "price"]

df = pd.read_csv(url, names=columns, na_values="?")
```

```python
# Convertir 'price' a numrico ignorando valores faltantes
df['price'] = pd.to_numeric(df['price'], errors='coerce')
```

```python
# Eliminar filas con valores NaN en 'price'
df.dropna(subset=['price'], inplace=True)
```

```python
# Estadsticas descriptivas bsicas
print("\nEstadsticas descriptivas de precios:")
print(df['price'].describe())
```

```python
# Determinar medidas de tendencia central
min_price = df['price'].min()
print(f"Mnimo: {min_price}")
```

```python
max_price = df['price'].max()
print(f"Mximo: {max_price}")
```

```python
mode_price = df['price'].mode()[0]
print(f"Moda: {mode_price}")
```

```python
median_price = df['price'].median()
print(f"Mediana: {median_price}")
```

```python
std_price = df['price'].std()
print(f"Desviacin estndar: {std_price}")
```

```python
# Determinar la cantidad ptima de bins usando la regla de Sturges
num_bins = int(1 + np.log2(len(df)))
```

```python
# Histograma de precios
plt.figure(figsize=(10,5))
sns.histplot(df['price'], bins=num_bins, kde=True, color='blue')
plt.title("Distribucin de Precios de los Automviles")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Diagrama de caja: Precio vs Tipo de Vehculo
plt.figure(figsize=(12, 5))
sns.boxplot(x='body-style', y='price', data=df)
plt.title("Relacin entre Precio y Tipo de Vehculo")
plt.xticks(rotation=45)
plt.show()
```

```python
# Diagrama de caja: Precio vs Tipo de Traccin
plt.figure(figsize=(12, 5))
sns.boxplot(x='drive-wheels', y='price', data=df)
plt.title("Relacin entre Precio y Tipo de Traccin")
plt.show()
```

```python
# Diagrama de caja: Precio vs Ubicacin del Motor
plt.figure(figsize=(12, 5))
sns.boxplot(x='engine-location', y='price', data=df)
plt.title("Relacin entre Precio y Ubicacin del Motor")
plt.show()
```
