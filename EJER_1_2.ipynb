# Lista de calificaciones

```python
from statistics import*

#lista de calificaciones del grupo a de probabilidad y estadistica

grupoa = [74, 80,65,90,100,95,45,80,73,20,100,96]
print(grupoa)

[70,70,65,90,100,95,45,80,73,20,100,96]

#La funcion type nos devuelve el tipo de dato de la variable
type(grupoa)

#sacar la media de las calificaciones con una formula

promedio = sum(grupoa) / len(grupoa)
print(promedio)

#utilizar la funcion mean que esta en la libreria statistics para sacar
#el promedio

promedio2 = mean(grupoa)
print(promedio2)

#Determine el valor de la mediana con la funcion median
mediana = median(grupoa)
print(mediana)

#Dtermine el valor de la moda con la funcion mode

moda = mode(grupoa)
print(moda)
```

# Tiempos de 100 metros

```python
#definimos una variable llamada tiempos100mts

tiempos100mts = [ 15.10,17.20,14.69,13.27,22.15,18.71,19.15,20.65,15.10,17.20]

print(tiempos100mts)

print("El tiempo promedio de los corredores es=>",mean(tiempos100mts))

mediana = median(tiempos100mts)
print("La mediana de los tiempos es=>",mediana)

moda = mode(tiempos100mts)
print("La moda de los tiempos es=>",moda)

multimodas = multimode(tiempos100mts)
print("La moda de los tiempos es=>",multimodas)
```

# Fabrica 1

```python
# Librería para generar numeros aleatorios
import random


fabrica1 = []
for i in range(100):
    numero_aleatorio = round(random.uniform(50, 100), 2)
    fabrica1.append(numero_aleatorio)

print(fabrica1)
media1 = mean(fabrica1)
mediana1 = median(fabrica1)
moda1 = mode(fabrica1)

print("Valores de la fabrica 1")
print("La media es= ",media1)
print("La medriana es= ",mediana1)
print("La moda es =",moda1)
```

# Fabrica 2

```python
# Librería para generar numeros aleatorios
import random


fabrica2 = []
for i in range(100):
    numero_aleatorio = round(random.uniform(50, 100), 2)
    fabrica2.append(numero_aleatorio)

print(fabrica2)
media2 = mean(fabrica2)
mediana2 = median(fabrica2)
moda2 = mode(fabrica2)

print("Valores de la fabrica 2")
print("La media es= ",media2)
print("La medriana es= ",mediana2)
print("La moda es =",moda2)
```

# Histrogramas Fabrica 1 y 2

```python
# librería de python para graficar valores
import matplotlib.pyplot as plt


plt.hist(fabrica1)
plt.title("Producción diaria de la fábrica 1")
```

```python
plt.hist(fabrica2)
plt.title("Producción diaria de la fábrica 2")
```

```python
fig, (ax1, ax2) = plt.subplots(nrows =1, ncols=2, figsize=(10, 5))
ax1.hist(fabrica1)
ax2.hist(fabrica2)
plt.show()
```

```python
plt.hist(fabrica1, bins=10, alpha=0.2, color='b', edgecolor = "black")
plt.axvline(media1, color='b', linestyle='dashed', linewidth=2, label='Media')
plt.axvline(mediana1, color='r', linestyle='dashed', linewidth=2, label='Mediana')
plt.axvline(moda1, color='y', linestyle='dashed', linewidth=2, label='Moda')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title("Producción diaria de la fabrica 1")
plt.legend()
plt.show()
```
